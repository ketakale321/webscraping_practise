import the library and use the constructor for bs4. The the page content and parse using html parser

```python
import bs4
soup_object = bs4.BeautifulSoup(page.content,"html.parser")
```

get all the interested element using the find_all method and class_ argument. This class is the class in the html

```python
news_lists =  soup_object.find_all("h4",class_="gPFEn")
print(news_lists)
```

traverse the list elements and get the cell content using .text property

```python
headlines = []
for news in news_lists:
    print(news.text)
```
